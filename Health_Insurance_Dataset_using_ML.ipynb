```python
#Step1: Gathering Data
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('/content/insurance.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
#step2: Data pre-processing where we will be using one hot encoding method
data['sex']=data['sex'].apply({'male':1,'female':0}.get)
```

```python
data.head()
```

```python
data['smoker']=data['smoker'].apply({'yes':1,'no':0}.get)
```

```python
data.head()
```

```python
data['region']=data['region'].apply({'southwest':0,'southeast':1,'northwest':2,'northeast':3}.get)
```

```python
data.head()
```

```python
#step3: Dividing data into dependent and independent data
x=data[['age','sex','bmi','children','smoker','region']]#dependent variables
```

```python
y=data[['expenses']]#independent variables
```

```python
#Splitting the data into training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
linear_model=LinearRegression()
```

```python
#training the model using fit()
linear_model.fit(x_train,y_train)
```

```python
y_pred=linear_model.predict(x_test)
```

```python
print(y_pred)#y_pred is the amount of health insurance as calculated by the ML model
```

```python
#checking accuracy of ML model
accuracy=linear_model.score(x,y)
```

```python
print('The accuracy of the ML model using linear Regression is',accuracy)
```
