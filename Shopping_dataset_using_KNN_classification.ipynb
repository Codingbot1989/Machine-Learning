```python
#Step1: Gathering Data
import numpy as np
import pandas as pd
```

```python
new=pd.read_csv('/content/Social_Network_Ads.csv')
```

```python
new.head()
```

```python
data=new.drop(['User ID'],axis=1)
```

```python
data.head()
```

```python
#step2: Data pre-processing where we will be using one hot encoding method
data['Gender']=data['Gender'].apply({'Male':1,'Female':0}.get)
```

```python
data.head()
```

```python
#step3: Dividing data into dependent and independent data
x=data[['Gender','Age','EstimatedSalary']]#independent variables
```

```python
y=data[['Purchased']]#dependent variables
```

```python
#Splitting the data into training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.75)#75% of dataset is used for training and 25% is used for testing
```

CREATING ML MODEL USING KNN ALGORITHIM

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)#n_neighbor is th value of k=5
```

```python
#training the model using fit()
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
```

```python
print(y_pred)
```

0 indicates person did not purchases the item and 1 indicates person purchased the item

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm=confusion_matrix(y_test,y_pred)
```

```python
acc=accuracy_score(y_test,y_pred)
```

```python
print(cm)
```
